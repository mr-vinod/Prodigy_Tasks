```python
import pandas as pd
import os
from matplotlib  import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/Google collab/population dataset.csv")
df=df.dropna()
df.head()
```

```python
countries = ['South Africa', 'Zimbabwe', 'Aruba', 'Afghanistan', 'India', 'China', 'Brazil', 'Sri Lanka', 'Russia', 'Turkey', 'Japan']
filter_data = df[df['Country Name'].isin(countries)]

plt.figure(figsize=(10, 6))
plt.title("Population of Specific Countries in 2002")
sns.barplot(x='Country Name', y='2002', data=filter_data)
plt.ylabel("Population")
plt.xlabel("Countries")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=filter_data, x="2002", kde=True, hue="Country Name", element="step")

plt.title("Distribution of 2002 Population with KDE by Country")
plt.xlabel("Population in 2002")
plt.ylabel("Frequency")

plt.legend(title="Country")
plt.show()
```

```python
years = ['1960', '1985', '2002', '2009', '2021']

plt.figure(figsize=(10,6))
for year in years:
    sns.histplot(data=filter_data, x=year, kde=True, hue="Country Name", element="step")

plt.title("Distribution of Population with KDE by Country Over the Years")
plt.xlabel("Population")
plt.ylabel("Frequency")

plt.legend(title="Country")
plt.show()
```

```python
df['population_change'] = df['1975'] - df['2002']

# Histogram
plt.figure(figsize=(10,6))
plt.hist(df['population_change'], bins=10, color='blue', edgecolor='black')
plt.title('Distribution of Population Change (1975-2002)')
plt.xlabel('Population Change')
plt.ylabel('Number of Countries')
plt.grid(True, which="both", ls="--", linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
df['percentage_increase'] = ((df['2002'] - df['2014']) / df['2014']) * 100
top_countries = df.nlargest(5, 'percentage_increase')
```

```python
bar_width = 0.35
index = range(len(top_countries))

plt.figure(figsize=(10, 8))
bar1 = plt.bar(index, top_countries['2015'], bar_width, label='2015', color='b', edgecolor='black')
bar2 = plt.bar([i+bar_width for i in index], top_countries['2015'], bar_width, label='2020', color='r', edgecolor='black')

plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population of Countries with Rapid Growth from 2015 to 2020')
plt.xticks([i+bar_width/2 for i in index], top_countries['Country Name'])
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df['percentage_increase'] = ((df['2000'] - df['2021']) / (df['2021'] + 1e-10)) * 100

bottom_countries = df.nsmallest(8, 'percentage_increase')

bar_width = 0.35
index = range(len(bottom_countries))

plt.figure(figsize=(10, 8))
bar1 = plt.bar(index, bottom_countries['2021'], bar_width, label='2021', color='b', edgecolor='black')
bar2 = plt.bar([i+bar_width for i in index], bottom_countries['2001'], bar_width, label='2001', color='r', edgecolor='black')

plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population of Countries with Least Growth from 2000 to 2021')
plt.xticks([i+bar_width/2 for i in index], bottom_countries['Country Name'])
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df['population_change_2015_2021'] = df['2015'] - df['2021']

country_with_largest_dip = df.nsmallest(5, 'population_change_2015_2021')

print(country_with_largest_dip[['Country Name', 'population_change_2015_2021']])
```

```python
country_data_1 = df[df['Country Name'] == 'India'].iloc[0]
country_data_2 = df[df['Country Name'] == 'China'].iloc[0]

years = ['1960', '1965', '1975', '1980', '2000', '2002', '2010', '2015', '2021']
population_values_1 = country_data_1[years]
population_values_2 = country_data_2[years]

bar_width = 0.35
index = range(len(years))

plt.figure(figsize=(15, 8))
bars1 = plt.bar(index, population_values_1, bar_width, label='India', color='blue', edgecolor='black')
bars2 = plt.bar([i + bar_width for i in index], population_values_2, bar_width, label='China', color='green', edgecolor='black')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of India vs China from 1960 to 2021')
plt.xticks([i + bar_width / 2 for i in index], years)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(years, population_values_1, marker='o', label='India', color='blue')
plt.plot(years, population_values_2, marker='o', label='China', color='red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Trend from 1960 to 2021')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
